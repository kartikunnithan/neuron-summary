```python
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
```

```python
df=pd.read_csv("/content/housing_data.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
y=df.pop("AboveMedianPrice")
x=df
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
model=Sequential()
model.add(Dense(10,input_dim=10,kernel_initializer='normal',activation='relu'))
model.add(Dense(1,kernel_initializer='normal',activation='relu'))
model.add(Dense(1,kernel_initializer='normal',activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model_history=model.fit(x_train,y_train,epochs=50,batch_size=10)
```

```python
model.summary()
```

```python
predictions=model.predict(x_test)
```

```python
pd.DataFrame({"Actual Values":y_test,"Predicted values":predictions.flatten()})
```

```python
mae=mean_absolute_error(y_test,predictions)
print("Model's accuracy:",mae)
```
